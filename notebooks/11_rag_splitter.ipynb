```python
from dotenv import load_dotenv
load_dotenv()

from langchain_text_splitters import RecursiveCharacterTextSplitter
texts = "Today, I want to speak from the heart about three pillars that define the future of our nation: our youth, our education system, and our farmers."

splitter = RecursiveCharacterTextSplitter(chunk_size=50, chunk_overlap=20)

chunks = splitter.split_text(texts)
chunks
```

### Document Text Loader

```python
from langchain_community.document_loaders import TextLoader, PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

textLoader = TextLoader("../data/speech.txt", encoding="utf-8")
pdfLoader = PyPDFLoader("../data/medical_report.pdf")

texts = textLoader.load()
pdfs = pdfLoader.load()

splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=20)
text_chunks = splitter.split_documents(texts)
pdf_chunks = splitter.split_documents(pdfs)

len(text_chunks)


pdf_chunks = splitter.split_documents(pdfs)
len(pdf_chunks)




```
