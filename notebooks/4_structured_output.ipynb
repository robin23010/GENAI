#### Getting structured output from llm

```python
from dotenv import load_dotenv
load_dotenv()
# these are library to create structured output using pydantic models
from pydantic import BaseModel, Field
from typing import List

from langchain_openai import ChatOpenAI
llm = ChatOpenAI(model="gpt-5-mini")

class Movie(BaseModel):
    name: str = Field(description="The movie's name")
    year: int = Field(description="The movie's release year")

class AllMovies(BaseModel):
    movies: List[Movie] = Field(description="A list of movies")


prompt = "List 5 movies released in the year 2000 or later, and provide their release year."
movies_llm= llm.with_structured_output(AllMovies)

response = movies_llm.invoke(prompt)
# convert the response to a dictionary and print it
print(response.model_dump())
```
