```python
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(model="gpt-4o")
prompt = [
    {"role": "system", "content": "You are a Javascript developer"},
    {"role": "user", "content": "Difference between map and filter ?"}
    ]
response = llm.invoke(prompt)
print(response.content)
```

###Prompt Template - Dynamic Prompts


```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

llm = ChatOpenAI(model="gpt-4o")
out = StrOutputParser()
prompt = ChatPromptTemplate.from_messages([
    {"role": "system", "content": "You are a Translator expert and translate the given text to {language} language"},
    {"role": "user", "content": "{query}"}
])
final_prompt = prompt.format(language="French", query="What is your name ?")
response = llm.invoke(final_prompt)
print(response.content)
```

```python
final_prompt = prompt.format(language="Hindi", query="What is your name ?")
response = llm.invoke(final_prompt)
print(response.content)
```

#####Chains

```python
chain = prompt | llm | out

res = chain.invoke({
    "language": "Hindi",
    "query": "What is your name?"
})

print(res)
```
